```python
import torch
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
 img = 'IMG-20240206-WA0002.jpg'
```

```python
 model
 results = model(img)
 results.print()
```

```python
 %matplotlib inline 
 plt.imshow(np.squeeze(results.render()))
 plt.show()
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
